```python
import pandas as pd
import seaborn as sns
```

```python
a = pd.read_csv('titanic_train.csv')
```

```python
a
```

```python
a.isnull()
```

```python
sns.heatmap(a.isnull())
```

```python
a.describe()
```

```python
sns.boxplot(x='Pclass',y='Age',data=a,palette='winter')
```

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
    
        if Pclass==1:
            return 37
        if Pclass==2:
            return 29
        
        else:
            return 24
        
    else:
        return Age
```

```python
a['Age']=a['Age'].fillna(method='ffill')
```

```python
a['Age'].describe()
```

```python
sns.heatmap(a.isnull())
```

```python
a.drop("Cabin",axis=1,inplace=True)
```

```python
a.head()
```

```python
sex=pd.get_dummies(a['Sex'],drop_first=True)
embark=pd.get_dummies(a['Embarked'],drop_first=True)
```

```python
a.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
a.drop(['PassengerId'],inplace=True,axis=1)
```

```python
a=pd.concat([a,sex,embark],axis=1)
```

```python
a.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=a.drop('Survived',axis=1)
y=a['Survived']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
predict=model.predict(X_test)

print(predict)
print(classification_report(y_test,predict))
print(confusion_matrix(y_test,predict))
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier()
model.fit(X_train,y_train)
predict = model.predict(X_test)

print(confusion_matrix(y_test,predict))
print(classification_report(y_test,predict))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()

model.fit(X_train,y_train)
predict=model.predict(X_test)
print(predict)
print(classification_report(y_test,predict))
print(confusion_matrix(y_test,predict))
```
