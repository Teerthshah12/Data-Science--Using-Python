```python
x=np.linspace(0,6,11)
y=x**2
#Create Figure (empty canvas)
fig=plt.figure()
#Add set of axes to figure
axes=fig.add_axes([0.01,0.01,0.3,0.3])#left,bottom,width,height(range 0to1)

#plot on that set of axes
axes.plot(x,y,'b')
axes.set_xlabel('set X Label')#Notice the use of set_ to begin methods
axes.set_ylabel('set Y Label')
axes.set_title('set Title')
```

```python
x=np.linspace(0,6,11)
y=x**2

fig=plt.figure()

axes1=fig.add_axes([0.01,0.01,0.5,0.5])
axes2=fig.add_axes([0.1,0.25,0.2,0.2])

axes1.plot(x,y,'b--')
axes1.set_xlabel('Set x label')
axes1.set_ylabel('Set y label')
axes1.set_title('Set title')

axes2.plot(y,x,'r*-')
axes2.set_xlabel('Set x label')
axes2.set_ylabel('Set y label')
axes2.set_title('Set title')
```

```python
fig1,ax=plt.subplots(2,3,sharex=True,sharey=True,layout='constrained')

ax[1,2].plot(x,y)
ax[1,2].set_title('Last')
ax[1,2].set_ylabel('y axis')
ax[0,1].plot(x,x)
```

```python
df=pd.DataFrame(np.random.randint(1,100,(4,3)),index=['Teerth','Naitik','Riya','Fenil'],columns=['Python','Data Science','DBMS'])
df
fig2, ax2=plt.subplots(2,2,layout='constrained')

ax2[0,0].plot(df.index,df.Python,'r-.',label='Python',lw='0.75',marker='s')
ax2[0,0].legend()
ax2[0,0].set_title('Python Marks')
ax2[0,0].set_xlabel('Marks')

ax2[0,1].plot(df.index,df.DBMS,'b--',label='DBMS',lw='0.25',marker='o')
ax2[0,1].legend()
ax2[0,1].set_title('DBMS Marks')

ax2[1,0].plot(df.index,df['Data Science'],'g-*',label='Data Science',marker='d')
ax2[1,0].legend()
ax2[1,0].set_title('Data Science Marks')

ax2[1,1].plot(df.index,df.Python,'r-.',label='Python',lw='1.25',marker='^')#'rgbcymk' is the colour red
ax2[1,1].plot(df.index,df.DBMS,'b--',label='DBMS',lw='1.25',marker='^')
ax2[1,1].plot(df.index,df['Data Science'],'g-*',label='Data Science',lw='1.25',marker='^')

ax2[1,1].legend()
ax2[1,1].set_title('All Marks')
```

Joint plot
jointplot() allows you to basically match up two distplots for bivariate data.With ypur choice of data what kind parameter to compare with
"scatter" 
"reg"
"resid"
"kde"
"hex"

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```
